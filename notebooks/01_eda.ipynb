```python
print(data["userid"].duplicated().any()) # Check if userid has any duplicated values

# Count gamerounds = 0
mask = data['sum_gamerounds'] == 0
print(f"Any True: {mask.any()}, Count of True: {mask.sum()}")

# Count gamerounds < 0
mask = data['sum_gamerounds'] < 0
print(f"Any True: {mask.any()}, Count of True: {mask.sum()}")

# Count gamerounds == 0 with retention_1 TRUE
mask = (data['sum_gamerounds'] == 0) & (data["retention_1"] == True) & (data["retention_7"] == False) 
print(f"Any True: {mask.any()}, Count of True: {mask.sum()}")

# Count gamerounds == 0 with retention_7 TRUE
mask = (data['sum_gamerounds'] == 0) & (data["retention_1"] == False)  & (data["retention_7"] == True) 
print(f"Any True: {mask.any()}, Count of True: {mask.sum()}") 

# Count gamerounds == 0 with both retention TRUE
mask = (data['sum_gamerounds'] == 0) & (data["retention_1"] == True) & (data["retention_7"] == True)
print(f"Any True: {mask.any()}, Count of True: {mask.sum()}") 
```
