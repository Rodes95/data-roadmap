```python
import sys
import sqlite3
import pandas as pd
import numpy as np

print("Python:", sys.version)
print("pandas:", pd.__version__)
print("numpy:", np.__version__)
```

```python
con = sqlite3.connect("data/db/test.db")

df = pd.DataFrame({
    "id": [1,2,3],
    "name": ["Alice", "Bob", "Charlie"],
    "age": [24, 31, 29]
})

df.to_sql("Test", con, if_exists="replace", index=False)

con.close()
```
